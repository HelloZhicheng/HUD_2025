```python
# Import necessary libraries
from IPython.display import Image
import requests
import pandas as pd

# Display all columns when showing dataframes
pd.options.display.max_columns = None

# Helper function: Convert JSON response to a pandas DataFrame
def json_to_dataframe(response):
    """
    Converts a JSON response from the Census API into a pandas DataFrame.
    The first row of the response is used as column names, and the rest as data.
    """
    return pd.DataFrame(response.json()[1:], columns=response.json()[0])
```

### Download for year 2010

```python
# URL for the first batch of Census data (2010 ACS 5-Year Estimates)
url2010_1 = (
    "https://api.census.gov/data/2010/acs/acs5?"
    "get=B01003_001E,B01001_002E,B01001_026E,B01001_020E,B01001_021E,B01001_022E,"
    "B01001_023E,B01001_024E,B01001_025E,B01001_044E,B01001_045E,B01001_046E,"
    "B01001_047E,B01001_048E,B01001_049E,B01001_006E,B01001_007E,B01001_008E,"
    "B01001_009E,B01001_010E,B01001_011E,B01001_012E,B01001_013E,B01001_014E,"
    "B01001_015E,B01001_016E,B01001_017E,B01001_018E,B01001_019E,B01001_030E,"
    "B01001_031E,B01001_032E,B01001_033E,B01001_034E,B01001_035E,B01001_036E,"
    "B01001_037E,B01001_038E,B01001_039E,B01001_040E,B01001_041E,B01001_042E,"
    "B01001_043E,B01001_003E,NAME&for=tract:*&in=state:26,39&in=county:*"
)

# Fetch and process the first batch of data
response2010_1 = requests.request("GET", url2010_1)
dataset2010_1 = json_to_dataframe(response2010_1)

# URL for the second batch of Census data (2010 ACS 5-Year Estimates)
url2010_2 = (
    "https://api.census.gov/data/2010/acs/acs5?"
    "get=B01001_004E,B01001_005E,B01001_006E,B01001_027E,B01001_028E,B01001_029E,"
    "B01001_030E,B05001_001E,B02001_002E,B02001_003E,B02001_004E,B02001_005E,"
    "B13002_001E,B11016_002E,B11016_003E,B11016_004E,B11016_005E,B11016_006E,"
    "B11016_007E,B11016_008E,B11016_010E,B11016_011E,B11016_012E,B11016_013E,"
    "B11016_014E,B11016_015E,B11016_016E,B19037_002E,B19037_019E,B19037_036E,"
    "B19037_053E,B09016_006E,B19131_056E,B19131_091E,B09001_002E,B09005_002E,"
    "B06009_002E,B06009_003E,B06009_004E,B06009_005E,B06009_006E,B14001_002E,"
    "B14001_010E,B06007_001E,B07003_004E,NAME&for=tract:*&in=state:26,39&in=county:*"
)

# Fetch and process the second batch of data
response2010_2 = requests.request("GET", url2010_2)
dataset2010_2 = json_to_dataframe(response2010_2)

# Merge the two datasets on common columns: NAME, state, county, and tract
dataset2010 = pd.merge(dataset2010_1, dataset2010_2, on=["NAME", "state", "county", "tract"])

# Convert 'state', 'county', and 'tract' columns to integer type for consistency
dataset2010 = dataset2010.astype({'state': 'int64', 'county': 'int64', 'tract': 'int64'})

# Load additional Census tract summary data from an Excel file
census_tract2010 = pd.read_excel('censustract_summary.xlsx', sheet_name="2010_2019")

# Merge the processed Census data with the Census tract summary data
df_combine2010 = pd.merge(dataset2010, census_tract2010, on=['state', 'county', 'tract'])

# Save the combined dataset as a CSV file for future use
df_combine2010.to_csv('datacombine2010.csv', index=False)

# Output a message indicating the script has completed
print("Census data for 2010 has been successfully processed and saved as 'datacombine2010.csv'.")
```

### Loop Function (download for year 2011-2020)

```python
# List of years to process
years = list(range(2013, 2020))

# The census tract summary data that you want to download
census_tract_data = pd.read_excel('censustract_summary.xlsx', sheet_name="2010_2019")

# Loop through each year and process data
for year in years:
    print(f"Processing data for year {year}...")
    
    # URLs for the two parts of Census data
    url_part1 = (
        f"https://api.census.gov/data/{year}/acs/acs5?"
        "get=B01003_001E,B01001_002E,B01001_026E,B01001_020E,B01001_021E,B01001_022E,"
        "B01001_023E,B01001_024E,B01001_025E,B01001_044E,B01001_045E,B01001_046E,"
        "B01001_047E,B01001_048E,B01001_049E,B01001_006E,B01001_007E,B01001_008E,"
        "B01001_009E,B01001_010E,B01001_011E,B01001_012E,B01001_013E,B01001_014E,"
        "B01001_015E,B01001_016E,B01001_017E,B01001_018E,B01001_019E,B01001_030E,"
        "B01001_031E,B01001_032E,B01001_033E,B01001_034E,B01001_035E,B01001_036E,"
        "B01001_037E,B01001_038E,B01001_039E,B01001_040E,B01001_041E,B01001_042E,"
        "B01001_043E,B01001_003E,NAME&for=tract:*&in=state:26,39&in=county:*"
    )
    
    url_part2 = (
        f"https://api.census.gov/data/{year}/acs/acs5?"
        "get=B01001_004E,B01001_005E,B01001_006E,B01001_027E,B01001_028E,B01001_029E,"
        "B01001_030E,B05001_001E,B02001_002E,B02001_003E,B02001_004E,B02001_005E,"
        "B13002_001E,B11016_002E,B11016_003E,B11016_004E,B11016_005E,B11016_006E,"
        "B11016_007E,B11016_008E,B11016_010E,B11016_011E,B11016_012E,B11016_013E,"
        "B11016_014E,B11016_015E,B11016_016E,B19037_002E,B19037_019E,B19037_036E,"
        "B06009_002E,B06009_003E,B06009_004E,B06009_005E,B06009_006E,B14001_002E,"
        "B14001_010E,B06007_001E,B07003_004E,NAME&for=tract:*&in=state:26,39&in=county:*"
    )
    
    # Fetch and process data from both URLs
    response_part1 = requests.request("GET", url_part1)
    response_part2 = requests.request("GET", url_part2)
    
    dataset_part1 = json_to_dataframe(response_part1)
    dataset_part2 = json_to_dataframe(response_part2)
    
    # Merge the two datasets
    dataset = pd.merge(dataset_part1, dataset_part2, on=["NAME", "state", "county", "tract"])
    
    # Convert 'state', 'county', and 'tract' columns to integer type
    dataset = dataset.astype({'state': 'int64', 'county': 'int64', 'tract': 'int64'})
    
    # Merge the Census data with the Census tract summary data
    combined_data = pd.merge(dataset, census_tract_data, on=['state', 'county', 'tract'])
    
    # Save the combined dataset as a CSV file
    output_filename = f"datacombine{year}.csv"
    combined_data.to_csv(output_filename, index=False)
    print(f"Data for year {year} saved as '{output_filename}'.\n")

print("All years processed successfully!")
```
